## EDA on `business` DataFrame

```python
state_dist = business.state.value_counts()
state_dist[state_dist > 1000].plot.bar()
plt.title("Number of Businesses by State");
```

```python
business.stars.plot.hist()
plt.title("Distribution of Star Ratings");
```

```python
cat = list()
for row in business.categories: 
    if row is None:
        continue
    for item in row.split(", "):
        cat.append(item)
cat = pd.Series(cat)
```

```python
over10000 = cat.value_counts()[cat.value_counts() > 1000]
print(f"The number of categories with over 10,000 categeries is:  {len(over10000)}")
over10000.plot.hist();
```

```python
df = business.copy()
for category in over10000.index: 
    df[category] = df.categories.str.contains(category)
```
